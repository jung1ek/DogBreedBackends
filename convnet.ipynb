```python
from transformers import AutoImageProcessor, AutoModelForImageClassification
import PIL
import requests
import torch
import os
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
image_processor = AutoImageProcessor.from_pretrained("./ConvNextV2-large-DogBreed")
model = AutoModelForImageClassification.from_pretrained("./ConvNextV2-large-DogBreed")
```

```python
# or using pipeline
# Use a pipeline as a high-level helper
from transformers import pipeline
pipe = pipeline("image-classification", model="dima806/133_dog_breeds_image_detection")
```

```python
model.to(device)
```

```python
model.eval()
```

```python
path = os.path.join("00d6e16493e6af2886292fa8823bdfe0.jpg")
image = PIL.Image.open(path).convert('RGB')
```

```python
inputs = image_processor(images=image, return_tensors="pt")
inputs.to(device)
outputs = model(**inputs)
logits = outputs.logits

# model predicts one of the 120 Stanford dog breeds classes
predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
